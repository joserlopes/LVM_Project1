```python
import matplotlib.pyplot as plt
import re

def get_times(file_path):
    with open(file_path) as fhand:
        all_lines = fhand.readlines()
    
    times = []
    for i in range(0, len(all_lines), 5):
        temp = []
        for j in range(i, i+2):
            temp.append(int(all_lines[j].split('=')[1]))
        for j in range(i+2, i+5):
            secs_str = re.findall("(.*)s", all_lines[j].split('=')[1])[0]
            temp.append(float(secs_str))

        times.append(temp)
    return times

# times = get_times('measurements/test')
len5_guess_inc = get_times('measurements/len_5_increasing_guesses')
len6_guess_inc = get_times('measurements/len_6_increasing_guesses')
len7_guess_inc = get_times('measurements/len_7_increasing_guesses')
len8_guess_inc = get_times('measurements/len_8_increasing_guesses')
len9_guess_inc = get_times('measurements/len_9_increasing_guesses')
guess20_inc_len = get_times('measurements/guess_20_incr_len')

def get_board_len(times_array):
    return [t[0] for t in times_array]

def get_guess_num(times_array):
    return [t[1] for t in times_array]

def get_build_time(times_array):
    return [t[2] for t in times_array]

def get_sat_time(times_array):
    return [t[3] for t in times_array]

def get_total_time(times_array):
    return [t[4] for t in times_array]
```

graph time - for board_len

```python
fig, ax = plt.subplots()

ax.plot(get_board_len(guess20_inc_len), get_build_time(guess20_inc_len), label='build formula time')
ax.plot(get_board_len(guess20_inc_len), get_sat_time(guess20_inc_len), label='first sat time')
ax.plot(get_board_len(guess20_inc_len), get_total_time(guess20_inc_len), label='all solutions time')

ax.set_xlabel('board len')
ax.set_ylabel('seconds')

# labels_board_len = [str(times[i][0]) for i in range(len(times))]
# ax.set_xticklabels(labels_board_len)
plt.title(label="Board with 20 guesses")
plt.legend()
plt.show()
```
